```python
def clean_data(file):
    cols = []
    for column in file.columns:
        cols.append(column.lower())
    file.columns = cols
    file = file.drop(['unnamed: 0'], axis=1)
    
    mode_vehicle_class = file['vehicle class'].mode()
    file['vehicle class'] = file['vehicle class'].fillna(mode_vehicle_class[0])
    mode_vehicle_size = file['vehicle size'].mode()
    file['vehicle size'] = file['vehicle size'].fillna(mode_vehicle_size[0])
    mode_vehicle_type = file['vehicle type'].mode()
    file['vehicle type'] = file['vehicle type'].fillna(mode_vehicle_type[0])
    
    file = file[file['state'].isna()==False] # Since these nulls are not a lot, we can filter them
    file = file[file['number of open complaints'].isna()==False] # Since these nulls are not a lot, we can filter them
    
    file['effective to date (date)'] = pd.to_datetime(file['effective to date'], errors='coerce')
    
    file = file.reset_index(drop = True)

    lst = file['effective to date (date)']
    idx = []
    for i in list(range(file.shape[0])):
        if lst[i].month < 3:
            idx.append(i)
    file = file[file.index == idx]
    display(file.shape[0])
```

```python
file = pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')
clean_data(file)
```
